```python
groudped_agg=df.groupby('date_of_record')['rainfall'].agg(['mean','sum','count'])
groudped_agg
```

```python
grouped_mean=df.groupby('season')['rainfall'].mean()
print(grouped_mean)
```

```python
grouped_mean=df.groupby('station_name')['rainfall'].sum().idxmax()
print(grouped_mean)
```

```python
grouped_mean=df.groupby('date_of_record')['rainfall'].sum().idxmax()
print(grouped_mean)
```

```python
grouped_mean=df.groupby('avg_temp')['rainfall'].mean()
print(grouped_mean)
```

```python
grouped_mean=df.groupby('month')['rainfall'].sum().idxmax()
print(grouped_mean)
```

```python
state_rain = df.groupby('state')['rainfall'].mean()
rain_heavy_states = state_rain[state_rain > state_rain.mean()]
print(rain_heavy_states)
```

```python
dry_states=state_rain[state_rain<state_rain.mean()]
print(dry_states)
```
